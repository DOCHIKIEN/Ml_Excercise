```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(42)
x = np.linspace(0, 10, 100)  
e = np.random.normal(0, np.sqrt(4), size=len(x))  
y = 2 * x + e

x = x.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(x,y)

y_pred = model.predict(x)
mse_linear = mean_squared_error(y_pred,y)
print(f'Mean Squared Error of Linear: {mse_linear:.2f}')
```

```python
poly_model = PolynomialFeatures(9)
x_poly = poly_model.fit_transform(x)

model = LinearRegression()
model.fit(x_poly,y)

y_pred_poly = model.predict(x_poly)
mse_poly = mean_squared_error(y_pred_poly, y)
print(f'Mean Squared Error of Polynominals: {mse_poly:.2f}')
```

```python
plt.scatter(x, y, color='blue', label='Data', alpha=0.8)
plt.plot(x, y_pred, color='red', label='Linear Regression')
plt.title('Linear Regression')
```

```python
plt.scatter(x, y, color='blue', label='Data', alpha=0.8)
plt.plot(x, y_pred_poly, color='orange', label='Poly Regresson')
plt.title('Polynominal Regression')
```
